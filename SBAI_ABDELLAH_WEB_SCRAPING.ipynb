```python
from msedge.selenium_tools import Edge, EdgeOptions

options= EdgeOptions()
options.use_chromium = True
driver = Edge(options=options)
url="https://www.mubawab.ma/fr/plr/casablanca-settat/listing-promotion:p:"
liens_temp=[]
date_pub_temp=[]

data={}

data["titre"]=[]
data["lien"]=[]
data["date_pub"]=[]
data["place"]=[]
data["prix_partir"]=[]
data["prix_m"]=[]
data["standing"]=[]
data["description"]=[]

for i in range(1,10): #9 pages
    
    driver.get(url+str(i))
    p = driver.find_element_by_id("listingUl") # ul id=listingUl
    promotions=p.find_elements_by_tag_name("li")#je prends li
    

    for i in promotions:
        date_pub=driver.find_element_by_css_selector("span.listingDetails.iconPadR").text
        
        #list_promo.append(i.get_attribute("linkref"))#affich lien
        #lis=i.get_attribute("linkref")
        liens_temp.append(i.get_attribute("linkref"))
        date_pub_temp.append(date_pub)
   
    data["lien"].extend(liens_temp)
    data["date_pub"].extend(date_pub_temp)
    
    for i in liens_temp:
        driver.get(i)#google ou edge revient les liens (liens_temp)
        
        try:
            data["titre"].append(driver.find_element_by_css_selector("h1.SpremiumH2").text)
        except:
            data["titre"].append("Not found!")
        
        try:
            data["place"].append(driver.find_element_by_css_selector("span.immoDetails.vAlignM.darkblue").text)
        except:
            data["place"].append("Not found!")
            
        try:    
            data["prix_partir"].append(driver.find_element_by_css_selector("h2.SpremiumH2.orangeText").text)
        except:
            data["prix_partir"].append("Not found!")
            
        try:
            data["prix_m"].append(driver.find_element_by_css_selector("span.SpremiumDetails").text)
        except:
            data["prix_m"].append("Not found!")

        try:
            data["standing"].append(list(map(lambda x: x.text, driver.find_elements_by_css_selector("div.promotionInfoBox.lHeight1.col-1.centered")))[:-1])
        except:
            data["standing"].append("Not found!")
        
        try:
            data["description"].append(driver.find_element_by_css_selector("p.changeDescrip").text)
        except:
            data["description"].append("Not found!")
    
    liens_temp=[]#afficher nouveau lien
    date_pub_temp=[]
```

```python
print(len(data["lien"]))
print(len(data["date_pub"]))

print(len(data["titre"]))
print(len(data["place"]))
print(len(data["prix_partir"]))
print(len(data["prix_m"]))
print(len(data["standing"]))
print(len(data["description"]))
```
